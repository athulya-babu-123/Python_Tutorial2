```python
import tkinter as tk
from tkinter import messagebox

CORRECT_USERNAME = "admin"
CORRECT_PASSWORD = "1234"

def login():
    username = entry_user.get().strip()
    password = entry_pass.get().strip()

    if username == "" or password == "":
        messagebox.showwarning("Missing Input", "Username and Password cannot be empty!")
        return

    if username == CORRECT_USERNAME and password == CORRECT_PASSWORD:
        messagebox.showinfo("Success", "Login Successful")
        entry_user.delete(0, tk.END)
        entry_pass.delete(0, tk.END)
    else:
        messagebox.showerror("Error", "Invalid Username or Password")
        entry_pass.delete(0, tk.END)
root = tk.Tk()
root.title("Login Form")
root.geometry("320x200")
root.resizable(False, False)

tk.Label(root, text="Username:").pack(pady=(20, 5))
entry_user = tk.Entry(root, width=30)
entry_user.pack()

tk.Label(root, text="Password:").pack(pady=(10, 5))
entry_pass = tk.Entry(root, width=30, show="*")
entry_pass.pack()

tk.Button(root, text="Login", width=12, command=login).pack(pady=20)

root.mainloop()
```
