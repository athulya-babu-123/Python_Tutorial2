```python
import tkinter as tk
from tkinter import messagebox

def generate_bill():
    try:
        price = float(entry_price.get())
        quantity = int(entry_quantity.get())

        total = price * quantity

        if total > 1000:
            discount = total * 0.10
        else:
            discount = 0

        final_amount = total - discount

        result_var.set(f"{final_amount:.2f}")

    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid numbers!")
        result_var.set("")

root = tk.Tk()
root.title("Billing System")
root.geometry("350x250")


tk.Label(root, text="Enter Item Price:").pack(pady=5)
entry_price = tk.Entry(root)
entry_price.pack(pady=5)


tk.Label(root, text="Enter Quantity:").pack(pady=5)
entry_quantity = tk.Entry(root)
entry_quantity.pack(pady=5)


tk.Button(root, text="Generate Bill", command=generate_bill).pack(pady=10)


tk.Label(root, text="Final Amount:").pack(pady=5)
result_var = tk.StringVar()
entry_result = tk.Entry(root, textvariable=result_var, state="readonly")
entry_result.pack(pady=5)

root.mainloop()
```
