```python
import tkinter as tk
from tkinter import messagebox
import re

def check_strength():
    password = entry_pass.get()

    if password == "":
        messagebox.showwarning("Input Error", "Password cannot be empty!")
        return

    length = len(password) >= 8
    has_digit = any(char.isdigit() for char in password)
    has_special = bool(re.search(r"[!@#$%^&*(),.?\":{}|<>]", password))

    score = sum([length, has_digit, has_special])

    if score == 1:
        result_var.set("Weak")
    elif score == 2:
        result_var.set("Moderate")
    elif score == 3:
        result_var.set("Strong")
    else:
        result_var.set("Weak")

root = tk.Tk()
root.title("Password Strength Checker")
root.geometry("350x200")

tk.Label(root, text="Enter Password:").pack(pady=10)
entry_pass = tk.Entry(root, show="*", width=30)
entry_pass.pack(pady=5)


tk.Button(root, text="Check Strength", command=check_strength).pack(pady=10)


tk.Label(root, text="Strength:").pack(pady=5)
result_var = tk.StringVar()
entry_result = tk.Entry(root, textvariable=result_var, state="readonly")
entry_result.pack(pady=5)

root.mainloop()
```
