```python
import tkinter as tk
from tkinter import messagebox

# Function to convert Celsius to Fahrenheit
def convert_temp():
    try:
        celsius = float(entry_celsius.get())
        fahrenheit = (celsius * 9/5) + 32
        
        # Show result with 2 decimal precision
        result_var.set(f"{fahrenheit:.2f}")
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter a valid number!")

# Create window
root = tk.Tk()
root.title("Temperature Converter")
root.geometry("300x200")

# Celsius Label
tk.Label(root, text="Enter Temperature (Celsius):").pack(pady=5)

# Celsius Entry
entry_celsius = tk.Entry(root)
entry_celsius.pack(pady=5)

# Convert Button
tk.Button(root, text="Convert to Fahrenheit", command=convert_temp).pack(pady=10)

# Result Label
tk.Label(root, text="Fahrenheit:").pack(pady=5)

# Read-only Result Field
result_var = tk.StringVar()
result_entry = tk.Entry(root, textvariable=result_var, state="readonly")
result_entry.pack(pady=5)

# Run the GUI
root.mainloop()
```
